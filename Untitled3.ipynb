```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Dropout, BatchNormalization, GlobalAveragePooling2D, Input
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing import image


# =========================
# CONFIG
# =========================
IMG_SIZE = 128
BATCH_SIZE = 16
EPOCHS = 20

TRAIN_DIR = "dataset/train"
VAL_DIR = "dataset/val"
TEST_DIR = "Test"

CLASSES = ["Clear", "Slime"]
EXT = (".jpg", ".jpeg", ".png")


# =========================
# DATAFRAME
# =========================
def build_df(root):
    data = []
    for cls in CLASSES:
        path = os.path.join(root, cls)
        if not os.path.isdir(path):
            print("Нет папки:", path)
            continue

        for f in os.listdir(path):
            if f.lower().endswith(EXT):
                data.append({
                    "filename": os.path.join(path, f),
                    "class": cls
                })

    return pd.DataFrame(data)


train_df = build_df(TRAIN_DIR)
val_df = build_df(VAL_DIR)

print(train_df["class"].value_counts())
print(val_df["class"].value_counts())


# =========================
# GENERATORS
# =========================
train_gen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=15,
    zoom_range=0.1,
    horizontal_flip=True
)

val_gen = ImageDataGenerator(rescale=1./255)

train = train_gen.flow_from_dataframe(
    train_df,
    x_col="filename",
    y_col="class",
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode="binary",
    classes=CLASSES
)

val = val_gen.flow_from_dataframe(
    val_df,
    x_col="filename",
    y_col="class",
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode="binary",
    classes=CLASSES,
    shuffle=False
)


# =========================
# MODEL (Sequential)
# =========================
model = Sequential([
    Input(shape=(IMG_SIZE, IMG_SIZE, 3)),

    Conv2D(32, 3, activation="relu", padding="same"),
    MaxPooling2D(),

    Conv2D(64, 3, activation="relu", padding="same"),
    MaxPooling2D(),

    Conv2D(128, 3, activation="relu", padding="same"),
    MaxPooling2D(),

    GlobalAveragePooling2D(),

    Dense(64, activation="relu"),
    Dropout(0.3),

    Dense(1, activation="sigmoid")
])

model.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)

model.summary()


# =========================
# TRAIN (без всякой херни)
# =========================
history = model.fit(
    train,
    validation_data=val,
    epochs=EPOCHS
)


# =========================
# PLOTS
# =========================
plt.plot(history.history["accuracy"], label="train")
plt.plot(history.history["val_accuracy"], label="val")
plt.legend()
plt.title("Accuracy")
plt.show()


# =========================
# TEST
# =========================
files = [f for f in os.listdir(TEST_DIR) if f.lower().endswith(EXT)]

plt.figure(figsize=(12, 6))

for i, name in enumerate(files):
    path = os.path.join(TEST_DIR, name)

    img = image.load_img(path, target_size=(IMG_SIZE, IMG_SIZE))
    arr = image.img_to_array(img) / 255.0
    arr = np.expand_dims(arr, 0)

    pred = model.predict(arr, verbose=0)[0][0]

    if pred > 0.5:
        label = "SLIME"
        color = "red"
    else:
        label = "CLEAR"
        color = "green"

    plt.subplot(3, 4, i+1)
    plt.imshow(img)
    plt.title(f"{label} ({pred:.2f})", color=color)
    plt.axis("off")

plt.tight_layout()
plt.show()
```
